```python
import pandas as pd
import numpy as np
```

```python
olympics=pd.read_csv("./olympics.csv")
olympics.head()
```

```python
olympics.head(50)
```

```python
print(olympics["Age"].isnull().sum())
olympics['Age'].fillna((olympics['Age'].mean()), inplace=True)
olympics.head()
```

```python
print(olympics["Age"].isnull().sum())
```

```python
print(olympics["Height"].isnull().sum())
olympics['Height'].fillna((olympics['Height'].mean()), inplace=True)
olympics.head()
```

```python
print(olympics["Weight"].isnull().sum())
olympics['Weight'].fillna((olympics['Weight'].mean()), inplace=True)
```

```python
print(olympics["Medal"].isnull().sum())
# leaving the medals null values the same, because the person might not get the medal for real
# so left the null values as it is
```

```python
olympics.to_csv("olympics_cleaned.csv")
```
