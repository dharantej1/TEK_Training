```python
import pandas as pd
import numpy as np
```

```python
flights=pd.read_csv("./Flights_Delay.csv")
flights=flights.loc[:,"ID":"CANCELLED"]
flights.head()
```

```python
flights.isnull().sum()
```

```python
print(flights["DEPARTURE_TIME"].isnull().sum())
flights['DEPARTURE_TIME'].fillna((flights['DEPARTURE_TIME'].mean()), inplace=True)
flights.head()
```

```python
print(flights["DEPARTURE_DELAY"].isnull().sum())
flights['DEPARTURE_DELAY'].fillna((flights['DEPARTURE_DELAY'].mean()), inplace=True)
flights.head()
```

```python
print(flights["TAXI_OUT"].isnull().sum())
flights['TAXI_OUT'].fillna((flights['TAXI_OUT'].mean()), inplace=True)
flights.head()
```

```python
print(flights["WHEELS_OFF"].isnull().sum())
flights['WHEELS_OFF'].fillna((flights['WHEELS_OFF'].mean()), inplace=True)
flights.head()
```

```python
print(flights["ELAPSED_TIME"].isnull().sum())
flights['ELAPSED_TIME'].fillna((flights['ELAPSED_TIME'].mean()), inplace=True)
flights.head()

print(flights["AIR_TIME"].isnull().sum())
flights['AIR_TIME'].fillna((flights['AIR_TIME'].mean()), inplace=True)
flights.head()

print(flights["WHEELS_ON"].isnull().sum())
flights['WHEELS_ON'].fillna((flights['WHEELS_ON'].mean()), inplace=True)
flights.head()

print(flights["TAXI_IN"].isnull().sum())
flights['TAXI_IN'].fillna((flights['TAXI_IN'].mean()), inplace=True)
flights.head()

print(flights["ARRIVAL_TIME"].isnull().sum())
flights['ARRIVAL_TIME'].fillna((flights['ARRIVAL_TIME'].mean()), inplace=True)
flights.head()

print(flights["ARRIVAL_DELAY"].isnull().sum())
flights['ARRIVAL_DELAY'].fillna((flights['ARRIVAL_DELAY'].mean()), inplace=True)
flights.head()
```
