```python
import matplotlib.pyplot as plt

# initializing the data
x = [1,8]
y = [3,0]

# plotting the data
plt.plot(x, y,"r^-")

# Adding the title
plt.title("Simple Plot")

# Adding the labels
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```
