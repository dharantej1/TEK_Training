```python
# isnull() returns the missing values as True and False 
autoprice_DF.isnull()
```

```python
# notnull() - reverse process of isnull() method
autoprice_DF.notnull()
```

```python
autoprice_DF.isnull().sum() # isnull() returns the True/ False values and we are summing up all those values and displaying
# symboling = 0 (No missing values), where 0 is the count of missing values
```

```python
autoprice_DF.isna().sum() # sum of missing rows present in each column
```
