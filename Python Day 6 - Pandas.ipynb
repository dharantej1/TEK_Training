#### concat()

```python
data_series={
    "Column 1":pd.Series([10,10,20,10,20,60,70]),
    "Column 2":pd.Series([10,10,20,10,20,60,70]),
    "Column 3":pd.Series([10,10,20,10,20,25,26]),
    "Column 4":pd.Series([10,10,20,10,20,60,70])
}
df=pd.DataFrame(data_series)

data_series1={
    "Column 1":pd.Series([10,10,20,10,20,60,70]),
    "Column 2":pd.Series([10,10,20,10,20,60,70])
}
df=pd.DataFrame(data_series)
df1=pd.DataFrame(data_series1)
print(pd.concat(df,df1))
```
