```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1=np.arange(0,5,0.1)
t2=np.arange(0,5,0.02)

plt.figure()
plt.subplot(231) # number of rows, cols, plot_number which ranges from 1 to n
plt.plot(t1,f(t1),"bo", t2,f(t2),"k")

plt.subplot(235)
plt.plot(t2,np.cos(2*np.pi*t2),"r--")

plt.subplot(233)
plt.plot(t2,np.sin(2*np.pi*t2),"b--")

plt.show()
```

```python
plt.figure(figsize=(6,4))

def f(x):
    return np.sin(x) - x*np.cos(x)
def g(x):
    """ derivative of g(x)"""
    return x*np.sin(x)

X=np.arange(-5,5,0.05)

fig,ax = plt.subplots(2, sharex="col", sharey="row")
print(fig,ax)
print(*ax)
print(type(ax[0]))

ax[0].plot(X, f(X), "r-", X, f(X), "k")
ax[0].set(title="The function f()")

ax[1].plot(X, g(X), "go", X, g(X), "k")
ax[1].set(xlabel="X Values", ylabel="Y Values", title="Derivative of function f()")


plt.show()
```

```python
x=np.linspace(0,10,300)
y=np.sin(x)+x

color='r'
linestyle='dashed'
linewidth=4

fig,ax=plt.subplots(figsize=(7,7))
fig.suptitle('Our first plot', fontsize=40)
ax.plot(x,y,c=color, linestyle=linestyle, linewidth=linewidth)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12,7))
fig.suptitle('Our first plot', fontsize=30)
ax.plot(x,y,c=color, linestyle=linestyle, linewidth=linewidth)
ax.set_ylabel('function', fontsize=30)
ax.set_xlabel('argument', fontsize=30)

for xtick in ax.xaxis.get_major_ticks():
    xtick.label.set_fontsize(18)
for ytick in ax.yaxis.get_major_ticks():
    ytick.label.set_fontsize(34)
plt.xticks(rotation=60)
plt.yticks(rotation=60)
plt.show()
```

```python
fig,axes=plt.subplots(1,3,figsize=(16,8))
ax1, ax2, ax3 = axes

x=np.linspace(0,10,500)

y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)

ax1.set_ylabel('sin(x)', fontsize=30)
ax1.set_xlabel('x', fontsize=30)

ax2.set_ylabel('cos(x)', fontsize=30)
ax2.set_xlabel('x', fontsize=30)

ax3.set_ylabel('tan(x)', fontsize=30)
ax3.set_xlabel('x', fontsize=30)

fig.tight_layout(pad=2)

ax1.plot(x, y1, c='r', linewidth=4)
ax2.plot(x, y2, c='g', linewidth=4)
ax3.plot(x, y3, c='b', linewidth=4)
plt.show()
```

```python
years=[str(i) for i in range(2016,2020)]
rice=[57,50,47,30]
wheat=[43,50,53,70]
paddy=[10,20,30,40]

plt.figure(figsize=(10,6))

plt.bar(years, rice, label="Rice")
plt.bar(years, wheat, bottom=rice, label="Wheat")
plt.bar(years, paddy, bottom=rice, label="Paddy")

plt.xlabel("Years", fontsize=14)
plt.ylabel("Preference (%)", fontsize=14)

plt.title("The result of rice/wheat export", fontsize=20)
plt.legend()
plt.show()
```
